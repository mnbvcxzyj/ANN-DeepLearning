```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 

# 피마 인디언 당뇨병 데이터셋 불러오기 
df = pd.read_csv("./pima-indians-diabetes3.csv")
```

```python
# heat map 
cmp = plt.cm.gist_heat

plt.figure() 
sns.heatmap(df.corr(), linecolor="white", annot=True, linewidths=0.2, vmax=0.7, cmap=cmp)
```

```python
import tensorflow as tf 
from tensorflow import keras
from keras import Sequential, Input
from keras.layers import Dense

# 관련 있는 정보들 저장 
x = df.iloc[:, 0:8] # 세부 정보 
y = df.iloc[:, 8] # 당뇨병 여부 
# 모델 설계 
model = Sequential() 
model.add(Input(shape=(8,))) # 입력층
model.add(Dense(12, activation="relu")) # 은닉층 1
model.add(Dense(8, activation="relu")) # 은닉층 2
model.add(Dense(1, activation="sigmoid")) # 출력층 

model.summary()
```

```python
# 모델 컴파일 
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# 모델 실행 
history = model.fit(x, y, epochs=100, batch_size=5, verbose=2)
```
