# 05. 선형회귀모델 - 먼저 긋고 수정하기 


## 경사 하강법 

```python
import numpy as np 
import matplotlib.pyplot as plt 

# 공부시간 
x = np.array([2,4,6,8])
# 성적 
y = np.array([81,93,91,97])

# 도화지 
plt.figure()
plt.scatter(x,y, marker='o', c='g', s=100)

# x축 라벨 달기 
plt.xlabel("Study Time(x)")

# y축 라벨 달기 
plt.ylabel("Score(y)")
plt.grid(True)
plt.show()  
```

```python
# 기울기 a 값과 y절편 b의 값 초기화 
a = 0 
b = 0 

# 학습률 정하기 
lr = 0.03 

# 몇 번 반복시킬지 횟수 
epoch = 2001 

# x값 길이 (데이터 개수)
n = len(x)
```

```python
# 경사하강법 (Gradient Descent)

for j in range(epoch): # 반복 횟수만큼 
    # 예측값 구하기 
    y_pred = a * x + b  
    # 에러 저장 
    error = y - y_pred 

    # a로 편미분 
    a_diff = (2/n) * sum(-x * error)

    # b로 편미분
    b_diff = (2/n) * sum(-error) 

    # 각 값에 학습률을 곱함 
    a = a - lr * a_diff 
    b = b - lr * b_diff 
    
    if (j%100)== 0:
        print("epoch=%.f, 기울기=%.4f, y절편=%.4f" % (j,a,b))
    
```

```python
# 최종 a값과 b값을 식에 대입하여 계산
y_pred = a*x+b 

# 도화지 
plt.figure()
plt.scatter(x,y, marker='o', c='g', s=100)
plt.plot(x, y_pred, marker="*", c="r", markersize=15
         )
# x축 라벨 달기 
plt.xlabel("Study Time(x)")

# y축 라벨 달기 
plt.ylabel("Score(y)")
plt.grid(True)
plt.show()  
```

## 다중 선형 회귀 

```python
# 공부 시간 
x1 = np.array([2,4,6,8])

# 과외 시간 
x2 = np.array([0,4,2,3])

# 성적 
y = np.array([81,93,91,97])

# 데이터 3d로 그리기 
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection="3d")
ax.scatter3D(x1,x2,y, s=200)
ax.set_xlabel("Study Time x1")
ax.set_ylabel("Private Lesson Time x2")
ax.set_zlabel("Score y")
plt.show()
```

```python
a1 = 0 
a2 = 0
b = 0 

# 학습률 설정 
lr = 0.01 

# 반복 횟수 
epoch = 2001

# x 개수 세기 
n = len(x)

# 경사하강법 
for j in range(epoch):  
    # 예측값 구하는 식  
    y_pred = a1 * x1 + a2 * x2 + b 

    # 실제값 - 예측값 오차 저장 
    error = y - y_pred 

    # 편미분하기 
    a1_diff = (2/n) * sum(-x1 * error)
    a2_diff = (2/n) * sum(-x2 * error)
    b_diff = (2/n) * sum(-error)


    # 학습률을 곱해 기존 기울기와 절편 업데이트 
    a1 = a1 - lr * a1_diff 
    a2 = a2 - lr * a2_diff 
    b = b - lr * b_diff 

    if (j%100)== 0:
        print("epoch=%.f, 기울기1=%.4f, 기울기2=%.4f, y절편=%.4f" % (j,a1, a2, b))
```

## 텐서플로우 사용하여 선형회귀, 다중 선형 회귀모델 구현

```python
import numpy as np 
import matplotlib.pyplot as plt 

from tensorflow.keras import Sequential, Input 
from tensorflow.keras.layers import Dense 

x = np.array([2,4,6,8])
y = np.array([81,93,91,97])

model = Sequential()

# 출력, 입력, 분석 방법에 따라 모델 설정 
model.add(Input(shape=(1,)))
model.add(Dense(1, activation="linear"))

# 경사 하강법 (SGD), 평균 제곱 오차 (MSE)
model.compile(optimizer="sgd", loss="mse")

# 오차 최소화 과정 2000번 반복 
model.fit(x,y, epochs=2000)

y_pred = model.predict(x)

# 그래프 그리기 
plt.figure()
plt.scatter(x,y,marker='o', c='g', s=100)
plt.plot(x, y_pred, marker="*", c='r', markersize=15)
plt.xlabel("study x")
plt.ylabel("score y ")
plt.grid(True)
plt.show() 

# 임의의 시간을 입력하여 학습 완료된 모델이 예측한 시험 점수 확인 
hour = 7 
prediction = model.predict([hour])
print(f"{hour}시간 공부한 학생의 예상 점수는 {prediction[0][0]:.2f}점 입니다.")
```

```python
import numpy as np 
import matplotlib.pyplot as plt 

from tensorflow.keras import Sequential, Input 
from tensorflow.keras.layers import Dense 

x = np.array([[2,0],[4,4],[6,2],[8,3]])
y = np.array([81,93,91,97])

model = Sequential()
model.add(Input(shape=(2,)))
model.add(Dense(1, activation="linear"))

model.compile(optimizer="sgd", loss="mse")
model.fit(x, y, epochs=2000)

hour = 7 
private_class = 3
result = model.predict([[hour,private_class]])

print("%.f시간을 공부를 하고, %.f시간의 과외를 받을 경우, 예상 점수는 %.f입니다." % (hour, private_class, result[0][0]))

# 그래프 그리기 
x1 = x[:,0]
x2 = x[:,1]

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(1,1,1,projection="3d")
ax.scatter3D(x1,x2,y,s=200)
ax.set_xlabel("StudyTime x1")
ax.set_ylabel("Private Lesson x2 ")
ax.set_zlabel("score Y")
plt.show()
```
