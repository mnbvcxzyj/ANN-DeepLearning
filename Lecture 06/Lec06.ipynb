# 로지스틱 회귀 모델 - 참 거짓 판단하기 
## 로지스틱 회귀 모델 -> 교차 엔트로피 오차 함수 

```python
import numpy as np
import matplotlib.pyplot as plt 
from tensorflow import keras 
from keras import Sequential, Input
from keras.layers import Dense 

x = np.array([2,4,6,8,10,12,14])
y = np.array([0]*3 + [1]*4)

model = Sequential() 
# 입력 1개 
model.add(Input(shape=(1,)))
# 시그모이드 함수 사용 
model.add(Dense(1, activation="sigmoid"))

# 교차 엔트로피 오차 함수 사용 
model.compile(optimizer="sgd", loss="binary_crossentropy")
model.fit(x,y, epochs=2000)

# 그래프 그리기 
plt.figure() 
plt.scatter(x,y,marker='o',c='k', s=100)
plt.plot(x,model.predict(x), marker='*', c='r', markersize=15)
plt.xlabel("study time x")
plt.ylabel("pass or failure")
plt.grid(True)
plt.show() 

# 임의의 학습 시간 넣기 
hour = 7 
prediction  = model.predict([hour])

print("%.f시간을 공부할 경우, 합격 예상 확률은 %.2f%%입니다." %(hour, prediction*100))
```
